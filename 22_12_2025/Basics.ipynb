```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, lit
spark = SparkSession.builder.appName("SparkByExamples.com").getOrCreate()
```

```python
data = [
    ("U001","Kochi",200),
    ("U002","Assam",200),
    ("U003","Kolkata",200),
    ("U004","Delhi",200),
    ("U005","Hydrabad",300),
]
```

```python
columns = ["user_id","city","amount"]
```

```python
df = spark.createDataFrame(data,columns)
```

```python
from pyspark.sql.functions import col,when
```

```python
df = df.withColumn("amount_with_tax",col("amount") * 1.10)
```

```python
df.show()
```

```python
df = df.replace("Delhi","New Delhi", subset=["city"])
```

```python
df.show()
```

```python
df = df.withColumn(
    "amount_category",
    when(col("amount") < 200, lit("Low"))
    .when(col("amount") < 400, lit("Medium"))
    .otherwise(lit("High"))
)

df.show()
```
