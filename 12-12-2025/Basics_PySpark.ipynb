```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
.appName('Basics') \
.getOrCreate()

spark.version
```

# First DataFrame

```python
data = [
    {"name":"Arjun","age":21},
     {"name":"Rahul","age":23},
     {"name":"John","age":19},
]

df = spark.createDataFrame(data)
df.show()
```

```python
df.printSchema()
```

```python
df.select('name').show()
df.select('name','age').show()
```

```python
df.filter(df.age > 20).show()
```

```python
from pyspark.sql.functions import col

df2 = df.withColumn('age',col('age')+5)
df2.show()
```

```python
from pyspark.sql.functions import lower

df3 = df.withColumn('name',lower(col('name')))
df3.show()
```

```python
df.drop("age").show()
```
